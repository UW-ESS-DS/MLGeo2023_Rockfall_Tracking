```python
from PIL import Image
import matplotlib.pyplot as plt

# define image path
image_path = 'frame0146masked.tif'

# Load the image
masked_image = Image.open(image_path)

# Display the binary mask
masked_image.show()

# Set the desired size of the displayed image
fig, ax = plt.subplots(figsize=(15, 15))

# Display the image using matplotlib
plt.imshow(masked_image)
plt.axis('on')  # Turn off axes if you don't want them
plt.show()
```

```python
import numpy as np

# Convert the image to a NumPy array
masked_array = np.array(masked_image)

# Print unique pixel values
print(np.unique(masked_array))

# Convert to binary using a threshold
threshold = 128  # Adjust this value based on your image characteristics
masked_image = masked_image.convert("L").point(lambda p: p > threshold and 255)

# Display the binary mask
masked_image.show()

# Set the desired size of the displayed image
fig, ax = plt.subplots(figsize=(15, 15))

# Display the image using matplotlib
plt.imshow(masked_image)
plt.axis('on')  # Turn off axes if you don't want them
plt.show()
```
